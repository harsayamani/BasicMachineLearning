```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
```

```
# import dataset
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```
# cek apakah ada nilai kosong
df.info()
```

```
# menghilankan kolom 'id' yang merupakan atribut tidak penting
data = df.drop(columns=['User ID'])
data = pd.get_dummies(data)
data
```

```
# memisahkan feature dan label
feature = data[['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']]
label = data['Purchased']
```

```
# membagi data menjadi train dan test set
feature_train, feature_test, label_train, label_test = train_test_split(feature, label, test_size=0.25, random_state=42) 
```

```
# membuat model logistic regression
model  = linear_model.LogisticRegression()
model.fit(feature_train, label_train)
```

```
#akurasi
model.score(feature_test, label_test)
```

```
#prediksi
model.predict([[49, 36000, 1, 0]])
```
