```
import sklearn
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from sklearn import tree
```

```
iris = datasets.load_iris()

data = iris.data
target = iris.target
```

```
# membuat model/klasifier decision tree
clsf = tree.DecisionTreeClassifier()
```

```
# evaluasi model
# model dikatakan baik jika berada di angka 0.85-0.99
scores = cross_val_score(clsf, data, target, cv=5)
scores
```
