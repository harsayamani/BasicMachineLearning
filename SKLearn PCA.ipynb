```
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import tree
```

```
# import dataset
iris = datasets.load_iris()
feature = iris.data
label = iris.target

# bagi dataset menjadi train set dan test set
feature_train, feature_test, label_train, label_test = train_test_split(feature, label, test_size=0.2)
```

```
# menghitung akurasi tanpa menggunakan PCA
decision_tree = tree.DecisionTreeClassifier()
first_model = decision_tree.fit(feature_train, label_train)
first_model.score(feature_test, label_test)
```

```
# menggunakan PCA
# membuat objek PCA dengan 4 principal component
pca = PCA(n_components=4)

# mengaplikasikan PCA pada dataset
pca_feature = pca.fit_transform(feature_train)

# melihat varian dari setiap atribut/feature
pca.explained_variance_ratio_
```

```
#melihat hasil dari variance kita bisa mengambil 2 principal component terbaik
pca = PCA(n_components=2)
feature_train_pca = pca.fit_transform(feature_train)
feature_test_pca = pca.fit_transform(feature_test)
```

```
second_model = decision_tree.fit(feature_train_pca, label_train)
second_model.score(feature_test_pca, label_test)
```
